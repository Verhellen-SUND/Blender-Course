## 1. Imports

We use `pandas` and `matplotlib`. If you prefer `polars` or `seaborn`, swap them in.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

DATA = Path('..') / 'data' / 'example.csv'
```

## 2. Load the example dataset

A small synthetic table of compound identifiers and IC50 values — just enough to demonstrate the toolchain. See `data/README.md` for the full metadata.

```python
df = pd.read_csv(DATA)
df
```

## 3. A first plot

IC50 distribution across the example compounds.

```python
fig, ax = plt.subplots(figsize=(6, 3.5))
ax.bar(df['compound'], df['ic50_nM'], color='#3c5e3e')
ax.set_ylabel('IC50 (nM)')
ax.set_title('Example compounds — IC50')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
